### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Att plotta i Matplotlib kan göras på två huvudsakliga sätt:

Objektorienterade (OO) gränssnittet:
Detta sätt innebär att man skapar figurer och axlar (figure och axis) manuellt, vilket ger dig mer kontroll över hur diagrammet ser ut. Detta sätt är oftast bäst när du vill göra mer avancerade eller komplexa visualiseringar.

Pyplot-gränssnittet (Stateful):
Detta sätt är enklare och liknar hur man gör diagram i MATLAB. Här används en global inställning som automatiskt hanterar figurer och axlar. Detta sätt är bra när du snabbt vill skapa enkla diagram.

Vilket sätt är bäst?
Det objektorienterade (OO) gränssnittet rekommenderas när du vill göra mer avancerade eller professionella diagram. Det ger dig bättre kontroll, flexibilitet och tydlighet, särskilt när du arbetar med flera diagram samtidigt. Men Pyplot-gränssnittet är användbart för snabba och enkla uppgifter.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Förklaring av begreppen i Matplotlib:
Figure (figur): Det är hela "ritningen" eller ytan där alla diagram och grafer skapas. En figur kan innehålla flera diagram (axes). Tänk på det som ett papper där du kan rita olika grafer.

Axes (axlar): Det är själva diagrammet inom figuren. Axes innehåller de två axlarna (x- och y-axeln) och själva datan som du vill visualisera. Varje figur kan ha flera axes (diagram).

Axis (axel): Det är själva linjerna som visar x- och y-värdena på diagrammet. Det finns en x-axel och en y-axel i varje axes, och dessa hjälper dig att förstå var datan ligger i diagrammet.

Artist (konstnär): Det är allt som du ser i en plot. Allt från linjer, text, och figurer till axlar och etiketter. Allt i Matplotlib som ritas på en figure är en "artist."

Kort sagt:

Figure = hela ritytan.
Axes = ett enskilt diagram i figuren.
Axis = x- och y-axlar i diagrammet.
Artist = alla visuella element i figuren.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
NumPy-arrayer: Matplotlib fungerar bäst med NumPy-arrayer, och det är den vanligaste datatypen som används för att plotta diagram. NumPy-arrayer kan vara i olika dimensioner, som 1D, 2D eller mer.

Exempel: np.array([1, 2, 3, 4])
Python-listor eller tupler: Du kan också använda vanliga Python-listor eller tupler. Dessa omvandlas automatiskt till NumPy-arrayer när de plottas.

Exempel: [1, 2, 3, 4] eller (1, 2, 3, 4)
Pandas DataFrames eller Series: Matplotlib integreras bra med Pandas. Det innebär att du kan plotta data direkt från en DataFrame eller Series.

Exempel: df['kolumn']
Andra datatyper: Skalära värden, dictionaries och range-objekt kan också användas, men dessa omvandlas ofta internt till arrayer.

Sammanfattning:
Den mest förväntade datatypen i Matplotlib är NumPy-arrayer, men du kan också använda Python-listor, tupler, och Pandas DataFrames eller Series.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data för scatter-plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data för bar-plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Skapa figur med 2 subplotter
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Första subplotten: scatter-plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.set_title("Scatter-Plot")
ax1.set_xlabel("Slumpmässigt X")
ax1.set_ylabel("Slumpmässigt Y")

# Andra subplotten: bar-plot
ax2.bar(names, values, color='orange')
ax2.set_title("Bar-Plot")
ax2.set_xlabel("Frukt")
ax2.set_ylabel("Mängd")

# Justera layout och visa plottarna
plt.tight_layout()
plt.show()
```
